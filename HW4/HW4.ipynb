```python
import torch

# Define a simple polynomial function: f(x) = x^2 - 6x + 5
def polynomial_function(x, a):
    temp = torch.tensor([0.0], requires_grad=True)
    for i in range(5) :
        temp = temp +  a[i]*(x**i)
    #print(a)
    #return a[0]*(x**0) + a[1]*(x**1) + a[2]*(x**2) + a[3]*(x**3) + a[4]*(x**4) 
    return temp 


def find_min(a) :
    # Create a tensor variable for the input
    x = torch.tensor([0.0], requires_grad=True)

    # Define the learning rate and the number of iterations
    learning_rate = 0.1
    epochs = 1000
    prev_x = float('inf')  # Initialize prev_x with a large value

    # Perform gradient descent without using an optimizer
    for epoch in range(epochs):
        # Calculate function value
        y_pred = polynomial_function(x,a)
        
        # Calculate gradients
        y_pred.backward()

        # Update x using gradients
        with torch.no_grad():
            x -= learning_rate * x.grad

            # Manually zero the gradients after updating
            x.grad.zero_()

        # Check for convergence by comparing consecutive x values
        if abs(prev_x - x.item()) < 1e-6:
            break
        
        # Update prev_x for convergence check in the next iteration
        prev_x = x.item()

    # Print the minimum value of the polynomial function and the corresponding x value
    min_x = x.item()
    min_value = polynomial_function(x,a).item()

    print(f"Minimum value of the function: {min_value} at x = {min_x}")
    
find_min([1,2,3,4,5])
```

```python
import os
from PIL import Image
from torchvision.datasets import VisionDataset
import torch
from torchvision import transforms

class CatDogdataset(VisionDataset):
    def __init__(self, root, train=False, transform=None):
        super().__init__(root, transform=transform)
        self.root=root
        self.train = train
        self.transform = transform
        self.images = []
        self.labels = []
        self.class_to_idx = {'cat': 0, 'dog': 1}  # Assuming 'cat' images are in cat folder and 'dog' in dog folder

        for dirname, _, filenames in os.walk(root):
            for filename in filenames:
                if filename.endswith('.jpg'):
                    file_path = os.path.join(dirname, filename)
                    self.images.append(file_path)
                    if train:
                        class_name = filename.split('.')[0]
                        self.labels.append(self.class_to_idx[class_name])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image_path = self.images[idx]
        image = Image.open(image_path).convert('RGB')  # Convert image to RGB
        if self.transform:
                image = self.transform(image)

        if self.train:
            label = self.labels[idx]

            
            return image, label
        else:
            return image

# Example usage:
transform = transforms.Compose([
    transforms.Resize((288, 288)), 
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(256)
    transforms.ToTensor(),  
])

transform_test = transforms.Compose([
    transforms.Resize((256, 256)), 
    transforms.ToTensor(),  
])

batch_size = 8

train_dataset = CatDogdataset(root='./HW1 image/reference', train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,shuffle=True)

test_dataset = CatDogdataset(root='./HW1 image/test', train=False, transform=transform_test)
testloader = torch.utils.data.DataLoader(test_dataset,batch_size=batch_size,shuffle=False)
```

```python
from torch.ao.nn.quantized.modules.batchnorm import BatchNorm2d
from torch.nn.modules.conv import Conv2d
import torch.nn as nn
import torch.nn.functional as F

class OurCNNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(3,64,kernel_size = 3,stride = 1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64,128,kernel_size=3,stride=2),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128,194,kernel_size=3,stride=2),
            nn.BatchNorm2d(194),
            nn.ReLU(inplace=True),
        )


        self.conv2 = nn.Sequential(
            nn.Conv2d(194,256,kernel_size=3,stride=2),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
        )
        
        self.conv3 = nn.Sequential(
            nn.Conv2d(256,512,kernel_size=3,stride=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
        )

        self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)

        # self.change_channel_1 = nn.Sequential(
        #     nn.Conv2d(256,512,kernel_size=1),
        #     nn.BatchNorm2d(512)
        # )


        # self.conv4 = nn.Sequential(
        #     nn.Conv2d(512,1024,kernel_size=3,stride=1,padding=1),
        #     nn.BatchNorm2d(1024),
        #     nn.ReLU(inplace=True),
        # )

        # self.change_channel_2 = nn.Sequential(
        #     nn.Conv2d(512,1024,kernel_size=1),
        #     nn.BatchNorm2d(1024)
        # )

        # self.conv5 = nn.Sequential(
        #     nn.Conv2d(1024,2048,kernel_size=3,stride=2,padding=1),
        #     nn.BatchNorm2d(2048),
        #     nn.ReLU(inplace=True),
        # )

        


        self.fc = nn.Sequential(
            nn.Linear(2*2*512,512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Linear(512,128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
            nn.Linear(128,2)
        )


    def forward(self, x):
        if not isinstance(x, torch.Tensor):
          x = torch.Tensor(x)
        out = self.conv1(x)
        out = self.maxpool(out)
        out = self.conv2(out)
        out = self.maxpool(out)
        out = self.conv3(out)
        out = self.maxpool(out)
        out = torch.flatten(out, 1)
        out = self.fc(out)
        return out
```

```python
device = torch.device('cuda')
model = OurCNNModel()
model.to(device)
```

```python
import torch.nn as nn
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=0.01)    
criterion = criterion.to(device)
```

```python
def train(input_data, model, criterion, optimizer):

    model.train()
    loss_list = []
    total_count = 0
    acc_count = 0
    for images, labels in input_data:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs, 1)
        total_count += labels.size(0)
        acc_count += (predicted == labels).sum().item()
        loss_list.append(loss.item())


    # Compute this epoch accuracy and loss
    acc = acc_count / total_count
    loss = sum(loss_list) / len(loss_list)
    return acc, loss
```

```python
max_epochs = 90
log_interval = 5 # print acc and loss in per log_interval time



# from torch.optim.lr_scheduler import ReduceLROnPlateau

# scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.1, patience=30)

train_acc_list = []
train_loss_list = []


torch.cuda.empty_cache()

for epoch in range(1, max_epochs + 1):



    train_acc, train_loss = train(trainloader, model, criterion, optimizer)


    train_acc_list.append(train_acc)
    train_loss_list.append(train_loss)


    torch.cuda.empty_cache()





    if epoch % log_interval == 0:
        print('=' * 20, 'Epoch', epoch, '=' * 20)
        print('Train Acc: {:.6f} Train Loss: {:.6f}'.format(train_acc, train_loss))
```

```python
def predict(input_data, model):
    model.eval()
    output_list = []
    with torch.no_grad():
        for images in input_data:
            images = images.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            output_list.extend(predicted.to('cpu').numpy().tolist())
    return output_list
```

```python
import torch.nn.functional as F
def predict_proba(input_data, model):
    model.eval()
    output_list = []
    labels_list = []
    prob_list = []
    with torch.no_grad():
        for images,labels in input_data:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            probs = F.softmax(outputs, dim=1)
            print("probs:",probs)
            _, predicted = torch.max(outputs.data, 1)
            prob_list.extend(probs.to('cpu').numpy().tolist())
            output_list.extend(predicted.to('cpu').numpy().tolist())
            labels_list.extend(labels.to('cpu').numpy().tolist())
    return prob_list, output_list, labels_list
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

prob_list, output_list, labels_list = predict_proba(trainloader, model)


cm = confusion_matrix(labels_list, output_list)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
# 從 prob_list 中提取正類別的概率
positive_probs = [prob[1] for prob in prob_list]  # 假設類別 1 是正類別

# 計算 AUROC
auroc = roc_auc_score(labels_list, positive_probs)
print("AUROC:",auroc)
```
